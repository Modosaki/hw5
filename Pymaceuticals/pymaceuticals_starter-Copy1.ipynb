```python
MD= pd.read_csv("data/mouse_drug_data.csv")
CTD= pd.read_csv("data/clinicaltrial_data.csv")
MCTD=MD.merge(CTD, on='Mouse ID', how="outer")
MCTD.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
#DrugTimeVolume:DTV
DTV=MCTD.groupby(['Drug','Timepoint'])
# Convert to DataFrame
DTVdf=pd.DataFrame(DTV['Tumor Volume (mm3)'].mean())
# Preview DataFrame
DTVdf=DTVdf.reset_index()
DTVdf.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
DTV=MCTD.groupby(['Drug','Timepoint'])
# Convert to DataFrame
DTVSEM=pd.DataFrame(DTV['Tumor Volume (mm3)'].sem())

# Preview DataFrame
DTVSEM=DTVSEM.reset_index()

DTVSEM.head(10)
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

# Convert to DataFrame

# Preview DataFrame
```

```python
DTVMO=MCTD.groupby(['Drug','Timepoint'])
# Convert to DataFrame
DTVdfMO=pd.DataFrame(DTVMO['Mouse ID'].count())
# Preview DataFrame
DTVdfMO=DTVdfMO.reset_index()
DTVdfMO.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview the Data Frame
```

```python

DrugMO=DTVdfMO.pivot(index='Timepoint',columns='Drug')['Mouse ID']
DrugMO.head()

#x=DrugMO.loc[0,['Capomulin']]
```

```python
# Generate the Plot (Accounting for percentages)
xs=DrugMO.index
x=DrugMO.loc[0,['Capomulin']]
x=x.sum()
q=DrugMO.loc[0,['Infubinol']]
q=q.sum()
w=DrugMO.loc[0,['Ketapril']]
w=w.sum()
e=DrugMO.loc[0,['Placebo']]
e=e.sum()
cps=DrugMO['Capomulin']/x*100

ips=DrugMO['Infubinol']/q*100

kps=DrugMO['Ketapril']/w*100
vps=DrugMO['Placebo']/e*100

fig3, ax = plt.subplots()

ax.errorbar(xs, cps, marker="v",color = 'red', alpha = 0.75)

ax.errorbar(xs, ips,marker="x",color = 'blue', alpha =0.75)

ax.errorbar(xs, kps,marker="o", color = 'green', alpha = 0.75)

ax.errorbar(xs, vps,marker="s", color = 'black', alpha = 0.75)



plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Sruvival Rate (%)")

ax.legend(loc="lower left")
plt.grid()
plt.show()

plt.savefig("../Images/survival.png")
# Save the Figure

# Show the Figure
plt.show()
```
